```python
first = ['о', '_', '_']
second = ['н', '_', '_']
third = ['о', '_', '_']
fourth = ['о','н', '_', '_']
fifth = ['ь']

result = []

for a in first:
    for b in second:
        for c in third:
            for d in fourth:
                for e in fifth:
                    word = a+b+c+d+e
                    if 'н' in word and 'о' in word:
                        result.append(word)
                        
letters = ['й','ц','ф','в','ч','ю','ж','э','ъ','х','з','щ','г']

result2 = []

for w in result:
    for l in letters:
        new_word = w.replace('_', l)
        result2.append(new_word)    
```
